```python
import re
```

```python
def is_valid_digit(c):
    if c.isdigit():
        return True
    if '\u0966' <= c <= '\u096F':
        return True
    return False
```

```python
def nepali_number(token):
    if token.startswith('+'):
        token_body = token[1:]
    else:
        token_body = token

    for c in token_body:
        if not is_valid_digit(c):
            return False

    # Convert Devanagari digits to Arabic
    dev_to_arabic = str.maketrans('०१२३४५६७८९', '0123456789')
    normalized = token_body.translate(dev_to_arabic)

    # Check valid patterns
    if normalized.startswith('977') and len(normalized) == 12 and normalized[3:5] == '98':
        return True        # 977XXXXXXXXXX
    elif normalized.startswith('98') and len(normalized) == 10:
        return True        # 98XXXXXXXX
    elif token.startswith('+') and normalized.startswith('977') and len(token) == 14:
        return True        # +977XXXXXXXXXX
    else:
        return False
```

```python
def email_validation(token):
  pattern = r'^[A-Za-z0-9](?:[A-Za-z0-9._+-]*[A-Za-z0-9])?@(?:[A-Za-z0-9-]+\.)+[A-Za-z]{2,4}$'
  return re.match(pattern, token) is not None
```

```python
def classify(token):
  if nepali_number(token):
    return "Phone Number"
  elif  email_validation(token):
    return "Email"
  else:
    return "Invalid"
```

```python

user_input = input("Enter tokens separated by spaces: ")
tokens = user_input.split()
for t in tokens:
    print(f"{t:25} -> {classify(t)}")
```
